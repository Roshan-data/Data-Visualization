Basic Plot Types in Seaborn
  1.Seaborn comes with built-in example datasets.
  2.These are small real-world datasets like restaurant tips, flight passenger counts, iris flower measurements, etc.
  3.They are mainly used for practice, examples, and learning plotting techniques without needing to manually download any data.
You can load these datasets directly into a pandas DataFrame using sns.load_dataset().

```python
#How to See All Available Datasets
import seaborn as sns
print(sns.get_dataset_names())
```

```python
#How to Load a Dataset
tips = sns.load_dataset('tips')
print(tips.head())
```

Why Seaborn Provides Datasets
 1.To quickly test different types of plots
 2.To learn plotting without needing your own data at first
 3.To create examples and tutorials easily
 4.To show real-world messy data handling (missing values, categorical data, etc.)

```python
import seaborn as sns
import matplotlib.pyplot as plt
#1. Line Plot
tips = sns.load_dataset('tips')
 
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot Example')
plt.show()
```

```python
#If we want scatter plot of same then
tips = sns.load_dataset('tips')
 
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Line Plot Example')
plt.show()
```

```python
#Bar Plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Average Bill per Day')
plt.show()
```

```python
#Box Plot
#Boxplots show distributions, medians, and outliers in one simple plot.

sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill per Day')
plt.show()
```

```python
#Working with Pandas DataFrames
#One of the biggest strengths of Seaborn:

import pandas as pd
 
df = pd.DataFrame({
    "age": [22, 25, 47, 52, 46, 56, 55, 60, 34, 43],
    "salary": [25000, 27000, 52000, 60000, 58000, 62000, 61000, 65000, 38000, 45000],
    "gender": ["M", "F", "M", "F", "F", "M", "M", "F", "F", "M"]
})
 
sns.scatterplot(x="age", y="salary", hue="gender", data=df)
plt.title('Salary vs Age Scatter Plot')
plt.show()
```
